# Plotting the 3 loss functions:

Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Setting up the actual and predicted values to be passed for input to the loss function

```python
y_pred = np.linspace(-1, 1, 1000)
y_act = np.zeros(1000)
```

# Squared Loss

```python
def sqLoss(x,y):
    res = (x - y) ** 2
    return res
```

```python
plt.plot(y_pred,sqLoss(y_act,y_pred))
plt.show()
```

# Absolute Loss

```python
def abLoss(x,y):
    res = (x - y)
    res = abs(res)
    return res
```

```python
plt.plot(y_pred,abLoss(y_act,y_pred))
plt.show()
```

# Cross-Entropy Loss

```python
y_act[:500].fill(1)
y_pred1 = np.linspace(0,1, 1000)
y_pred1[0]= 0.01
y_pred1[999]= 0.99
```

```python
def logLoss(y_pred1, y_act):
    cc = np.array([])
    for i in range(len(y_act)):
        if y_act[i] == 1:
            cc = np.append(cc,-np.log2(y_pred1[i]))
        else:
            cc = np.append(cc,-np.log2(1 - y_pred1[i]))
    return cc
```

```python
plt.plot(y_pred1,logLoss(y_pred1, y_act))
plt.show()
```
